Membaca image dengan Open CV

```python
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread('image/yoga.jpeg')
cv2_imshow(img)
```

Mengubah image menjadi hitam putih / meng-gray scale image

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_img)
```

Membaca Histogram dari gray scale image

```python
import matplotlib.pyplot as plt

histogram = cv2.calcHist([gray_img], [0], None, [256], [0, 256])

plt.plot(histogram)
plt.show()
```

Menghitung Color Histogram (Histogram berwarna)

```python
for i, col in enumerate(['b', 'g', 'r']):
    hist = cv2.calcHist([img], [i], None, [256], [0, 256])
    plt.plot(hist, color = col)
    plt.xlim([0, 256])

plt.show()
```
